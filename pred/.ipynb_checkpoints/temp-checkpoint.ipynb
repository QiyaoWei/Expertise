## Expertise and Randomness

```python
cblind_palete = sns.color_palette("colorblind", as_cmap=True)
ce = []
cfr = []
tar = []
ipw = []
paths = ["0.1.txt", "1.0.txt", "2.0.txt", "3.0.txt", "4.0.txt"]
# paths = ["propscale0.01numlayers0lossmult1.txt", "propscale0.1numlayers0lossmult1.txt", "propscale1numlayers1lossmult1.txt", "propscale2.0lossmult1000scalefactor1.0predscale1.0.txt", "propscale3.0lossmult1000scalefactor1.0predscale1.0.txt"]
for i in range(len(paths)):
    ce.append([])
    cfr.append([])
    tar.append([])
    ipw.append([])
    flag = False
    with open("0.01False" + paths[i], "r") as myfile:
        for line in myfile:
            if flag:
                flag = False
                temp = line.split("ce")
                temp = temp[-1].split("cfr")
                ce[i].append(float(temp[0]))
                temp = temp[-1].split("tar")
                cfr[i].append(float(temp[0]))
                temp = temp[-1].split("ipw")
                tar[i].append(float(temp[0]))
                ipw[i].append(float(temp[-1]))
            if "seed" in line:
                flag = True

xaxis = range(len(paths))
plt.figure()
ceval = np.mean(np.array(tar) - np.array(ce), axis=1)
ceerr = scipy.stats.sem(np.array(tar) - np.array(ce), axis=1)
# cfrval = np.array([-0.03982956,-0.06827717,-0.08099458,-0.08319345,-0.08399985])
# cfrerr = np.array([0.00184486,0.00229917,0.0027645,0.00297836,0.00369757])
cfrval = np.mean(np.array(tar) - np.array(cfr), axis=1)
cfrerr = scipy.stats.sem(np.array(tar) - np.array(cfr), axis=1)
plt.plot(xaxis, ceval, label="ce", color=cblind_palete[0], linestyle="-", marker="d")
plt.plot(xaxis, cfrval, label="cfr", color=cblind_palete[1], linestyle="--", marker="o")
plt.fill_between(xaxis, ceval-ceerr, ceval+ceerr, alpha=0.1, color=cblind_palete[0])
plt.fill_between(xaxis, cfrval-cfrerr, cfrval+cfrerr, alpha=0.1, color=cblind_palete[1])
plt.legend()
plt.xlabel("Prop Scale")
plt.ylabel("PEHE (TAR-Method)")
labels = ['0.1', '1', '2', '3', '4']
plt.xticks(xaxis, labels)
plt.title("Expertise and Randomness")
plt.savefig("propscale.pdf")
plt.show()
```

```python
ce = []
cfr = []
ipw = []
tar = []
with open("propscale1.0lossmult1scalefactor1.0predscale1.0.txt", "r") as myfile:
    for line in myfile:
        if "seed" in line:
            ce.append(list())
            cfr.append(list())
            ipw.append(list())
            tar.append(list())
            seed = int(line.split("seed: ")[-1])
        else:
            temp = line.split("ce")
            temp = temp[-1].split("cfr")
            ce[seed-1].append(float(temp[0]))
            temp = temp[-1].split("tar")
            cfr[seed-1].append(float(temp[0]))
            temp = temp[-1].split("ipw")
            tar[seed-1].append(float(temp[0]))
            ipw[seed-1].append(float(temp[-1]))

ipwval = np.flip(np.mean(np.array(tar) - np.array(ipw), axis=0))
ipwerr = np.flip(scipy.stats.sem(np.array(tar) - np.array(ipw), axis=0))
xaxis = range(len(ipwval))

plt.figure(figsize=(12,8))
plt.plot(xaxis, ipwval, label="ipw", color=cblind_palete[2], linestyle=":", marker="^")
plt.fill_between(xaxis, ipwval-ipwerr, ipwval+ipwerr, alpha=0.1, color=cblind_palete[2])
plt.legend()
plt.xlabel("Increasing Expertise")
plt.ylabel("Reward")
plt.title("Expertise and Covariates")
plt.savefig("feature.pdf")
plt.show()
```
